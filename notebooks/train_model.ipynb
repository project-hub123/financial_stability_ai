## 1. Импорт библиотек

```python
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

## 2. Загрузка исходных данных

```python
data_path = "../data/financial_data.csv"
df = pd.read_csv(data_path)

df.head()
```

## 3. Формирование финансовых коэффициентов

```python
def safe_divide(a, b):
    return np.where(b == 0, 0, a / b)

features = pd.DataFrame()

features["current_ratio"] = safe_divide(
    df["current_assets"], df["current_liabilities"]
)

features["equity_ratio"] = safe_divide(
    df["equity"], df["total_assets"]
)

features["return_on_assets"] = safe_divide(
    df["profit"], df["total_assets"]
)

features.head()
```

## 4. Разделение выборки

```python
X = features
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

## 5. Обучение модели Random Forest

```python
model = RandomForestClassifier(
    n_estimators=300,
    max_depth=6,
    random_state=42
)

model.fit(X_train, y_train)
```

## 6. Оценка качества модели

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy модели: {accuracy:.2%}")

print("\nОтчет классификации:")
print(classification_report(y_test, y_pred))
```

## 7. Сохранение обученной модели

```python
model_path = "../models/model1.pkl"
joblib.dump(model, model_path)

model_path
```
